```python
stopwords = nltk.corpus.stopwords.words('English')
def text_preprocessor(doc):
    doc = doc.lower().strip()
    try:
        doc = doc.replace('.',". ")
        doc = doc.replace(',',", ")
        doc = doc.replace('!',"! ")
    except:
        pass
    doc = re.sub(r"[^a-z\s]","",doc) #to rmv space and anything not lower
#     doc = [d.strip() for d in doc.split() if d not in stopwords]
#     doc = " ".join(doc)
    return doc
```

```python
chatbot_df['pattern'].apply(text_preprocessor)
```

```python
chatbot_df['Cleaned_pattern'] = chatbot_df['pattern'].apply(text_preprocessor)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
y = le.fit_transform(chatbot_df['tag'])
```

```python
y
```

```python
# X.toarray() # bcos of floats num we cant have an embedding layer
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(num_words = 3000)
tokenizer.fit_on_texts(chatbot_df['Cleaned_pattern'])
X = tokenizer.texts_to_sequences(chatbot_df['Cleaned_pattern'])
```

```python
X
```

```python
len(tokenizer.word_index)
```

```python
X = pad_sequences(X,maxlen = 300)
vocab_size = len(tokenizer.word_index) + 1 
```

```python
from tensorflow.keras.layers import LSTM,Flatten,Dense,Input,Embedding
from tensorflow.keras import Sequential
```

```python
model = Sequential()
model.add(Embedding(input_dim=vocab_size,output_dim = 200 ,input_length =300))
model.add(LSTM(32,return_sequences = True))
model.add(LSTM(64,return_sequences = True))
model.add(Flatten())
model.add(Dense(38,activation = "softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer="adam",metrics = ["accuracy"])
```

```python
model.fit(X,y,epochs = 50,validation_split=0.15,verbose = 1)
```

```python
pd.Series("How are you today")
```

```python
inp = "How are you today"
input_token = tokenizer.texts_to_sequences(pd.Series(inp))
inp_token = pad_sequences(input_token,maxlen = 300)

y_pred = model.predict(inp_token)
```

```python
y_pred
```

```python
predicted_tag = le.inverse_transform([np.argmax(y_pred)])[0]
possible_responses = tag_responses[predicted_tag]
```

```python
responses = np.random.choice(possible_responses)
```

```python
responses
```

```python
def return_responses(inp):
    input_token = tokenizer.texts_to_sequences(pd.Series(inp))
    inp_token = pad_sequences(input_token,maxlen = 300)
    probabilities = model.predict(inp_token)
    y_pred = np.argmax(probabilities)
    tag = le.inverse_transform([y_pred])[0]
    print(tag)
#     possible_responses = tag_responses[tag]
    return np.random.choice(tag_responses[tag])
```

```python
while True:
    user_input=input("What is on your mind?: ")
    print(f"user: {user_input} ")
    
    reply = return_responses(user_input)
    
    print(f"Bot: {reply}")
    
```
