```python
df = pd.DataFrame()
for website in websites:
    website_df = pd.read_csv(f"{path}/{website}",delimiter ='\t',header = None)
    df = pd.concat([df,website_df],axis = 0)
```

```python
df.columns = ['Review','Sentiment']
```

```python
def text_preprocessor(doc):
    doc = doc.lower().strip()
    try:
        doc = doc.replace('.',". ")
        doc = doc.replace(',',", ")
        doc = doc.replace('!',"! ")
    except:
        pass
    doc = re.sub(r"[^a-z\s]","",doc) #to rmv space and anything not lower
    doc = [d.strip() for d in doc.split() if d not in stopwords]
    doc = " ".join(doc)
    return doc
```

```python
stopwords = nltk.corpus.stopwords.words('English')
```

```python
df['Cleaned_review']=df['Review'].apply(text_preprocessor)
```

```python
embedding_matrix = np.zeros((token_size,vec_size))
for word,index in tokenizer.word_index.items():
    if word in word_vec.wv:
        embedding_matrix[index] = word_vec.wv[word]
```

# RNN

```python
from tensorflow.keras.layers import Embedding, Dense, Dropout, LSTM, SimpleRNN,Flatten
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
model.add(Embedding(input_dim = token_size,output_dim = vec_size,input_length = vec_size,
                    weights = [embedding_matrix]))
model.add(LSTM(units=256,dropout = 0.3,recurrent_dropout = 0.2,return_sequences = True))
model.add(LSTM(units=128,return_sequences = True))
model.add(Flatten())
model.add(Dense(units=1,activation = 'sigmoid'))
```

```python
y.value_counts()
```

```python
model.compile(loss='binary_crossentropy',optimizer = 'adam',metrics =['accuracy'])
```

```python
model.fit(X_train,y_train,epochs= 30,validation_split = 0.15,verbose = 1)
```

```python
model.evaluate(X_test,y_test)
```
