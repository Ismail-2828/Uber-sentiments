# Text preprocessing using Nltk

```python
nltk.download('stopwords') #list of stopwords in nltk library
nltk.download('punkt') #for tokennization
```

```python
stopwords = nltk.corpus.stopwords.words('English')
```

```python
def text_preprocessor(doc):
    doc = doc.lower().strip()
    doc = re.sub(r"[^a-z\s]","",doc) #to rmv space and anything not lower
    doc = [d for d in doc.split() if d not in stopwords]
    return doc
```

```python
df['Review'].apply(text_preprocessor)
```

```python
def text_preprocessor(doc):
    doc = doc.lower().strip()
    try:
        doc = doc.replace('.',". ")
        doc = doc.replace(',',", ")
        doc = doc.replace('!',"! ")
    except:
        pass
    doc = re.sub(r"[^a-z\s]","",doc) #to rmv space and anything not lower
    doc = [d.strip() for d in doc.split() if d not in stopwords]
    doc = " ".join(doc)
    return doc
```

```python
X= df['Cleaned_review']
y = df['Sentiment']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
```

```python
vectorizer.fit(X)
```

```python
X = vectorizer.transform(X)
```

```python
X.shape
```

```python
vectorizer.get_feature_names_out()
```

```python
X.toarray()
```

```python
pd.DataFrame(X.toarray(),columns = vectorizer.get_feature_names_out())
```

```python
from sklearn.model_selection import train_test_split
```

```python
y.value_counts()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train,y_train)
```

```python
lr_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,lr_model.predict(X_test))
```
